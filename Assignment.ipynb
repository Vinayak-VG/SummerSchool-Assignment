```python
N=1000
```

```python
Y=np.random.randint(0,2,N) # uniiform disrtibution {0,1} integers only
```

```python
Y_P=np.random.rand(N) # uniform distribution [0,1) 
```

```python
CS=np.sum(Y*np.log2(Y_P)+(1-Y_P)*np.log2(1-Y_P))*(-1/N)
```

```python
print('Cross Enropy Is',CS)
```

```python
class pairfind:
  
  def __init__(self, s, anw):
    self.result = {}
    k = 1
    for i in range(len(s)):
      for j in range(len(s)):
        if s[i] + s[j] == anw:
          self.result[k] = [i, j]
          k += 1
    print(self.result)
```

```python
b=pairfind([4,10,6,14,18,19,20],20)
```
