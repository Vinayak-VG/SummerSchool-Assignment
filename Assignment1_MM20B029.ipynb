```python
class MyClass:
 x=np.array([10,20,10,40,50,60,70])
 y=50


 def get_indices(self):
  k=1
  i=0
  l=len(self.x)
  output={}

  while i<l:
      j=0
      while j<l:
        if j!=i:
         if self.x[i]+self.x[j]==self.y:
          output[k]=[i,j]
          k=k+1
        j=j+1
      i=i+1
  print(output)


w=MyClass()
w.get_indices() 
```
