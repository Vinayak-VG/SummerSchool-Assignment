```python
import numpy as np
import random as rnd
```

```python
n=100
```

```python
y=[]
i=0
while i<n:
    y.append(rnd.randint(0,1))
    i+=1
```

```python
y_cap=[]
i=0
while i<n:
    y_cap.append(rnd.random())
    i+=1
```

```python
print(y)
```

```python
print(y_cap)
```

```python
s=0
i=0
while i<n:
    s+=(y[i]*np.math.log(y_cap[i],2))+((1-y[i])*np.math.log(1-y_cap[i],2))
    i+=1
O=(-1/n)*s
```

```python
print(O)
```

```python
lst=[]
i=0
while i<100:
    lst.append(rnd.randint(0,100))
    i+=1
```

```python
print(lst)
```

```python
targ=rnd.randint(0,200)
print(targ)
```

```python
class q_2:
    def __init__(self,num,target):
        self.num=num
        self.target=target
    def find(self):
        k=1
        index={}
        for i in range(len(self.num)):
            for j in range(len(self.num)):
                if(self.num[i]+self.num[j]==self.target and i!=j):
                    index[k]=i,j
                    k+=1
        return index
```

```python
objct=q_2(lst,targ)
print(objct.find())
```
